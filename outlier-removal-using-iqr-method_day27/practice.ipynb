```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import scipy.stats as stats 
import seaborn as sns
```

```python
df = pd.read_csv('breast_cancer.csv')
df
```

```python
for cols in df.columns: 
    plt.figure(figsize=(18,4))
    plt.subplot(131)
    sns.histplot(df[cols],kde=True)
    plt.subplot(132)
    stats.probplot(df[cols],plot=plt)
    plt.subplot(133)
    sns.boxplot(df[cols])
    
```

```python
percentile25 = df['mean area'].quantile(0.25)
percentile75 = df['mean area'].quantile(0.75)
```

```python
IQR = percentile75 - percentile25
```

```python
upper_limit = percentile75 + 1.5* IQR
lower_limit = percentile25 - 1.5* IQR
```

```python
print("Upper limit",upper_limit)
print("Lower limit",lower_limit)
```

```python
df[df['mean area'] > upper_limit]
```

## Cappping 
### here i do capping i don't do trimming becouse trmming is easy just delete data i only do capping

```python
new_df = df.copy()
new_df["mean area"] = np.where(
        new_df['mean area']>upper_limit,
        upper_limit,
        np.where(
            new_df['mean area']< lower_limit,
            lower_limit,
            new_df['mean area']
        )
    )
```

```python
 
plt.figure(figsize=(18,8))
plt.subplot(2,2,1)
sns.histplot(df['mean area'],kde=True)
plt.subplot(2,2,2)
stats.probplot(new_df['mean area'],plot=plt)
plt.subplot(2,2,3)
sns.boxplot(new_df['mean area'])
plt.subplot(2,2,4)
sns.boxplot(df['mean area'])
```
