```python
import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OrdinalEncoder , LabelEncoder
```

```python
df = pd.read_csv('Student_Performance.csv',usecols=['final_grade','parent_education','internet_access','school_type','gender'])
df['final_grade'].unique()
# df
```

```python
x_train , x_test, y_train , y_test = train_test_split(df.drop('final_grade',axis=1),
                                                      df['final_grade'],
                                                      test_size=0.3, random_state=0)
```

```python
x_train
```

### WHen categories permeter are important when in some column order are very imp

#### if i don't give 'categories' permeter in OrdinalEncoder then this give automatically random order you don't know about the order 
#### it is important to give 'categories permeter'

```python
oe = OrdinalEncoder(categories=[['male','female','other'],
                                ['private','public'],
                                ['phd','post graduate','graduate','diploma','high school','no formal'],
                                ['yes','no']])
```

```python
oe.fit(x_train)
```

```python
X_train = oe.transform(x_train)
```
