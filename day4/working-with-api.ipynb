```python
import pandas as pd 
import requests
```

```python
api_key = 'fa47aef278ef9b6ffdc228dcf66455fd'
df = pd.DataFrame()
for i in range(1,10):  
    url = f'https://api.themoviedb.org/3/movie/top_rated?api_key={api_key}&language=en-US&page={i}'
    res = requests.get(url)
    data = res.json()
    temp = pd.DataFrame(data['results'], columns=['id','title','popularity','release_date','vote_average','vote_count'])
    df = pd.concat([df,temp],ignore_index=True)
# data
```

```python
df.to_csv('movie.csv')
```
