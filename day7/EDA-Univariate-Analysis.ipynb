```python
import pandas as pd 
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
```

# Catogorical Data

### a. CountPlot 

```python
sns.countplot(x='Pclass',data=df)

# sns.countplot(x='Survived',data=df)
```

```python
df['Pclass'].value_counts().plot(kind='pie',autopct='%.2f')
```
