## Capping

```python
df.columns
```

```python
up_lo_cols = ['mean texture','mean smoothness','mean symmetry','worst smoothness']
up_lo_cols
```

```python
upper_and_lower  = {}
for cols in up_lo_cols:
    upper_and_lower[f"{cols}_upper"] = df[cols].mean() + 3*df[cols].std()
    upper_and_lower[f"{cols}_lower"] = df[cols].mean() - 3*df[cols].std()

upper_and_lower
```

```python
for cols in up_lo_cols:
    df[cols] = np.where(
        df[cols]>upper_and_lower[f'{cols}_upper'],
        upper_and_lower[f'{cols}_upper'],
        np.where(
            df[cols]<upper_and_lower[f'{cols}_lower'],
            upper_and_lower[f'{cols}_lower'],
            df[cols]
        )
    )
```
