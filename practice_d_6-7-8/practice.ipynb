```python
df = pd.read_csv('train.csv',dtype={'Ticket':str})
df
```

```python
# prof = ProfileReport(df)
# prof.to_file(output_file='output.html')

df['Family'] = df['SibSp'] + df['Parch']
df
```

```python
# i Want to create a Family as catagorial data like if in family peoples are 
# less then 1 then type  alone in column
# if 2 or 3 then small family 
# if people greater then 4 or 4 big famill 
df['Family'] = df['Family'].apply(lambda x: 
    (
    "SmallFamily" if x  == 2 or x == 3 else
    "BigFamily" if x >= 4 else 
    "Alone" 
    ))
df
```
