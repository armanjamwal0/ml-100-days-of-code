```python
import pandas as pd 
import numpy as np 
from  sklearn.impute import SimpleImputer
from sklearn.preprocessing import OrdinalEncoder , OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('used_cars_data.csv',usecols=['Name','Location','Mileage','Owner_Type','Transmission','Fuel_Type','Price','Kilometers_Driven'])
# df['Location'].value_counts()
```

```python
x_train , x_test , y_train, y_test = train_test_split(df.drop('Kilometers_Driven',axis=1),
                                                      df['Kilometers_Driven'],
                                                      test_size=0.3,
                                                      random_state=0)
# x_train.shape
x_train
```

```python
tranfomer = ColumnTransformer(transformers=[
    ('tnf1',SimpleImputer(),['Price']),
    ('tnf2',OrdinalEncoder(categories=[['Fourth & Above','Third','Second','First']]),['Owner_Type']),
    ('tnf3',OneHotEncoder(sparse_output=False,drop='first'),['Location','Fuel_Type'])
],remainder='passthrough')
```

```python
feature_names = tranfomer.get_feature_names_out()
tra = tranfomer.fit_transform(x_train)
pd.DataFrame(tra,columns=feature_names)
# print(feature_names)
```
