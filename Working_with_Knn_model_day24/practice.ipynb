```python
import pandas as pd 
import numpy as np

from sklearn.model_selection import train_test_split 
from sklearn.impute import  KNNImputer , SimpleImputer 
from sklearn.linear_model import LogisticRegression ,LinearRegression
from sklearn.preprocessing import LabelEncoder ,StandardScaler
from sklearn.metrics import accuracy_score

from sklearn.metrics import r2_score
from sklearn.dummy import DummyClassifier
```

```python
df = pd.read_csv('csv/train.csv',usecols=['Age',"Fare",'Survived'])
df.isna().mean()*100
df
```

```python
X = df.drop('Survived',axis=1)
Y = df['Survived']
```

#### without , random_state=42 you get  data every time when u run code diffrent shuffle data with , random_state=42 you get same shuffle you can try 

```python
x_train , x_test , y_train , y_test = train_test_split(X, Y , test_size=0.2 , random_state=42)
x_train
```

### i use here base line and baseline means "No brain model score "

### base line is more important 
### Always check baseline before trusting accuracy.

```python
dummy = DummyClassifier(strategy="most_frequent")
dummy.fit(x_train,y_train)
print("Baseline accuracy:", dummy.score(x_test, y_test))
```

```python
score = []
for i in range(1,25):
    Knn = KNNImputer(n_neighbors=i)
    x_train_trf = Knn.fit_transform(x_train)
    x_test_trf = Knn.transform(x_test)
    
    # scale = StandardScaler()
    # x_train_s = scale.fit_transform(x_train_trf)
    # x_test_s = scale.transform(x_test_trf)
    
    m = LogisticRegression()
    m.fit(x_train_trf,y_train)
    y_pred = m.predict(x_test_trf)

    score.append(accuracy_score(y_test,y_pred))
    
```
