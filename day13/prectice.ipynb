```python
import pandas as pd 
import numpy as np 
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.read_csv('used_cars_data.csv',usecols=['Name','Location','Mileage','Owner_Type','Transmission','Fuel_Type','Price'])
df['Price'] = df['Price'].fillna(0.0)
df['Fuel_Type']
```

```python
df['Location'].value_counts()
```

## 3. OneHotEncoding using Sklearn


```python
x_train , x_test, y_train , y_test = train_test_split(df.drop('Price',axis=1),df['Price'],test_size=0.3,random_state=0)
x_train
```

```python
ohe = OneHotEncoder(drop='first',sparse_output=False, dtype=np.int32)
x_train_stake = ohe.fit_transform(x_train[['Fuel_Type','Owner_Type']])
x_test_stake = ohe.transform(x_test[['Fuel_Type','Owner_Type']])

x_train_stake
```
